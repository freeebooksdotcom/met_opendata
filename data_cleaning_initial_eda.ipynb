```python
def rename_columns(column_name):
    new_name = ""
    for letter in column_name:
        if letter.isupper():
            new_name+=letter.lower()
        elif letter == ' ':
            new_name+='_'
        else:
            new_name+=letter
    return new_name
```

```python
df.rename(columns=rename_columns, inplace=True)
```
