```python
def rename_columns(column_name):
    new_name = ""
    for letter in column_name:
        if letter.isupper():
            new_name+=letter.lower()
        elif letter == ' ':
            new_name+='_'
        else:
            new_name+=letter
    return new_name
```

### Exploring Artwork tags

```python
unique_tags=[]
for tag in list(df.tags.unique())[1:]:
    unique_tags+=tag.split('|')
unique_tags=set(unique_tags)
```

```python
unique_tags
```

```python
def turn_tags_into_lists(messy_tags):
    return str(messy_tags).split('|')
```

```python
tags=df.tags.map(turn_tags_into_lists)
```

```python
tags
```

```python
### Code to find most popular tags
count_dict={}
for work in tags:
    for tag in work:
        if tag in count_dict.keys():
            count_dict[tag]+=1
        else:
            count_dict[tag]=1

```
